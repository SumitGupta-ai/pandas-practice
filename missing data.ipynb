```python
import numpy as np
import pandas as pd
```

Finding missing Data

```python
data = {
    "A": [1, 2, np.nan, 4],
    "B": [2, 3, 5, np.nan],
    "C": [3, 2, np.nan, 9],
    "D": [8, np.nan, 7, np.nan]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

```python
df
```

Filling the missing Data

```python
df.fillna(9)
```

```python
value = {"A": 120,"B": 97, "C": 789,"D": 102}
df.fillna(value=value)
```
