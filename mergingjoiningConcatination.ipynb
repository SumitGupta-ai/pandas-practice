```python
import numpy as np
import pandas as pd
```

merging 2 dataframes

```python
employees = pd.DataFrame({
    "Employees_id" : [1,2,3,4,5],
    "name" : ['rahul','nitish','shubham','deleep','sonu'],
    "departement" : ["Hr",'IT','sbi','ubi','hr'],
    
})

salaries = pd.DataFrame({
    'Employees_id' : [1,2,3,6,7],
    'salary' : [30000,45333,646664,345464,23435],
    'bonus' : [400,2333,5344,3234,5232],
    
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='Employees_id',how='inner')
```

```python
pd.merge(employees,salaries,on='Employees_id',how='outer')
```

```python
pd.merge(employees,salaries,on='Employees_id',how='left')
```

```python
pd.merge(employees,salaries,on='Employees_id',how='right')
```

Concatination 2 DataFrame

```python
df = pd.DataFrame({
    'A' : ["A0","A1","A2"],
    'B' : ['B0','B1','B2'],
    'C' : ['C0','C1','C2'],
    
})

df1 = pd.DataFrame({
    'A': ['A3','A4','A5'],
    'B' : ['B3','B4','B6'],
    'C' : ['C3','C4','C5']
})
```

```python
df
```

```python
df1
```

```python
pd.concat([df,df1])
```

```python
pd.concat([df,df1],axis=1)
```

**joinging 2 data frames**

```python
df1 = pd.DataFrame({
    'name': ['Manish','Aman','Shivam']
    
},index = [1,2,3])

df2 = pd.DataFrame({
    'score': [90,56,76]
},index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1,how='outer')
```
