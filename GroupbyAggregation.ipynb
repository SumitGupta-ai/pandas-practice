```python
import numpy as np
import pandas as pd
```

### Groupby

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Salesperson': ['Amit', 'Neha', 'Ravi', 'Pooja', 'Amit', 'Ravi', 'Neha', 'Amit'],
    'Product': ['Laptop', 'Mobile', 'Laptop', 'Tablet', 'Tablet', 'Laptop', 'Mobile', 'Mobile'],
    'Sales': [20000, 15000, 18000, 22000, 25000, 17000, 16000, 14000],
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Mar', 'Mar']
}

df = pd.DataFrame(data)
df
```

```python
summ = df.groupby('Product')['Sales'].sum()
summ
```

```python
df.groupby('Salesperson')['Sales'].sum()
```

```python
df.groupby(['Salesperson','Product'])['Sales'].sum()
```
